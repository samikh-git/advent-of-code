```python
def create_matrix(file):
    matrix = []
    for l in file:
        row = []
        for x in l.replace("\n", ""):
            row.append(x)
        matrix.append(row)
    return matrix
```

```python
# Part 1
def count_neighbors(matrix):
    select = set()
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == "@":
                bad_numbers = -1
                for t in [-1, 0, 1]:
                    for l in [-1, 0, 1]:
                        if i + t >= 0 and i + t < len(matrix) and j + l >= 0 and j + l < len(matrix[i]):
                            bad_numbers += int(matrix[i + t][j+l] == "@")
                if bad_numbers < 4:
                    select.add((i, j))
    return len(select)

def solution(input):
    with open(input) as f:
        m = create_matrix(f)
        return count_neighbors(m)
```

```python
#Part 2
def alter(matrix):
    select = set()
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == "@":
                bad_numbers = -1
                for t in [-1, 0, 1]:
                    for l in [-1, 0, 1]:
                        if i + t >= 0 and i + t < len(matrix) and j + l >= 0 and j + l < len(matrix[i]):
                            bad_numbers += int(matrix[i + t][j+l] == "@")
                if bad_numbers < 4:
                    select.add((i, j))

    for p in select:
        matrix[p[0]][p[1]] = "x"

    return len(select)

def solution(input):
    counter = 0
    with open(input) as f:
        m = create_matrix(f)
    while True:
        to_remove = alter(m)
        if not to_remove:
            return counter
        counter += to_remove
```
