# Advent of Code Day 2 - Gift Shop

Link to the problem statement: [https://adventofcode.com/2025/day/2](https://adventofcode.com/2025/day/2)

```python
def check_is_pattern_of_2(s):
    pattern = s[:len(s)//2]
    return pattern*2 == s
```

```python
def is_repeated_pattern(string):
    pattern = ""
    for s in string[:-1]:
        pattern += s
        if pattern*(len(string)//len(pattern)) == string:
            return True
    return False
```

```python
def solution(pattern_f):
    with open("input.txt") as f:
        line = f.readline().split(",")

        invalid_ids = []

        for li in line:
            l = li.split("-")
            a, b = int(l[0]), int(l[1])
            for i in range(a, b+1):
                s = str(i)
                if pattern_f(s):
                    invalid_ids.append(int(s)) 
        
        return sum(invalid_ids)
```

```python
#part 1
print("The solution for part 1 is:", solution(check_is_pattern_of_2))

#part 2
print("The solution for part 2 is :", solution(is_repeated_pattern))
```
