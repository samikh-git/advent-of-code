# Advent of Code Day 1 - Unlocking the safe

Link to the problem statement: [https://adventofcode.com/2025/day/1](https://adventofcode.com/2025/day/1)

Modular arithmetic to the rescue!

```python
#Part 1
cross_zeroes = 0
dial = 50

with open("input.txt") as f:
    for line in f:
        number = int(line[1:])
        if line[0] == "L":
            dial = (dial - number) % 100
        elif line[0] == "R":
            dial = (dial + number) % 100
        if dial == 0:
            cross_zeroes += 1

print(cross_zeroes)
```

```python
#Part 2
cross_zeroes = 0
dial = 50

with open("input.txt") as f:
    for line in f:
        number = int(line[1:])
        for _ in range(number):
            if line[0] == "L":
                dial = (dial - 1) % 100
            elif line[0] == "R":
                dial = (dial + 1) % 100
            if dial == 0:
                cross_zeroes += 1

print(cross_zeroes)
```
