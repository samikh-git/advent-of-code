## Part 1

```python
def find_max_joltage(s):
    max = -1
    for i in range(len(s)):
        for j in range(i + 1, len(s)):
            if int(s[i] + s[j]) > max:
                max = int(s[i] + s[j])
    return max
```

```python
with open("input.txt") as f:
    sum1 = 0
    for l in f:
        sum1 += find_max_joltage(l)

print(sum1)
```

## Part 2

```python
def combinations(s, build):
    if s == "" or build == 0:
        return [""]
    if len(s) == build:
        return [s]
    without_ = combinations(s[1:], build)
    without_best = max(without_, key = int)[0]
    if without_best and  without_best[0] <= s[0]:
        with_ = combinations(s[1:], build - 1)
        return [s[0] + x for x in with_] + list(without_)
    return without_
```

```python
with open("input.txt", "r") as f:
    sum2 = 0
    i = 0
    for l in f:
        l = l.replace("\n", "")
        sum2 += int(max(combinations(l, 12), key = int))
        i += 1
        print("Completed ", i, " operations")

print(sum2)
```
